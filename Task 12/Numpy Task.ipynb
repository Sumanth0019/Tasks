```python
import numpy as np
```

N-Dimension Arrays


```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
```

```python
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

```python
a=np.zeros((2,3))
b=np.ones((2,3))
c=np.arange(0,12,3)

print(a)
print(b)
print(c)
```

Indexing & slicing

```python
arr=np.array([[10,20,30],[40,50,60]])

print(arr)
print(arr[0,1])
print(arr[:,1])
print(arr[0:2,1:3])
```

array operations

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print(a+b)
print(a*b)
print(a**2)
```

Broadcasting

```python
a=np.array([[1,2,3],[4,5,6]])
b=np.array([10,20,30])

print(a+b)
```

```python
angles=np.array([0,np.pi/2,np.pi])
print(angles)

print(np.sin(angles))
print(np.cos(angles))
print(np.tan(angles))
```

Statistical functions

```python
arr=np.array([1,2,3,4,5])

print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
```

Shape manipulation

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr)

print(arr.reshape(3,2))
print(arr.flatten())
print(arr.ravel())
print(arr.T)
```

Joining and splitting arrays

```python
a=np.array([[1,2]])
b=np.array([[3,4]])

print(np.concatenate((a,b),axis=0))
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
arr = np.array([1,2,3,4,5,6])

print(np.array_split(arr, 3))
```
