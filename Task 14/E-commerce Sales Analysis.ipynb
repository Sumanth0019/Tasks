```python
import numpy as np
```

## Generating Dataset

```python
np.random.seed(0)

days = 30
products = 10

# sales quantity per day per product
sales = np.random.randint(0, 50, size=(days, products))

print(sales[:5])
```

adding prices

```python
prices = np.random.randint(100, 1000, size=products)

print(prices)
```

revenue calc

```python
revenue = sales * prices

total_revenue_per_day = np.sum(revenue, axis=1)
total_revenue_per_product = np.sum(revenue, axis=0)
```

best sales day

```python
best_day = np.argmax(total_revenue_per_day)
print("Best day:", best_day)
```

best selling product

```python
best_product = np.argmax(total_revenue_per_product)
print("Best product:", best_product)
```

worst selling product

```python
worst_product = np.argmin(total_revenue_per_product)
print(worst_product)
```

Avg daily revenue

```python
avg_daily_revenue = np.mean(total_revenue_per_day)
print(avg_daily_revenue)
```

Top 3 Products

```python
top_products = np.argsort(-total_revenue_per_product)[:3]
print(top_products)
```

High demand days

```python
high_sales_days = np.where(total_revenue_per_day > avg_daily_revenue)
print(high_sales_days)
```

```python
import matplotlib.pyplot as plt

plt.plot(total_revenue_per_day)
plt.title("Daily Revenue Trend")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.show()
```
