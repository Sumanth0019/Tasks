```python
import numpy as np
```

##### Generating the data

```python
np.random.seed(42)
data = np.random.randint(10, 100, size=100)
print("100 days data from sensor:", data)
```

##### using indexing & Slicing

```python
print("First element",data[0])
print("From 1 to 15 element",data[0:25])
print("Last element",data[-1])
print("every 2nd value",data[::2])
```

##### copy vs view

```python
copy_data=data.copy()
view_data=data.view()

#updating new data
data[0]=99
print(copy_data[0])
print(view_data[0])
```

##### boolean masking

```python
#finding reading abve 70
high_values=data[data>70]
print(high_values)
```

##### math ops

```python
sqrts=np.sqrt(data)
print(sqrts**2)
```

##### adding random noise

```python
data = data + np.random.randint(-10, 10, size=100)

print(data)
```

##### cleaning the data

```python
data = data[(data >= 5) & (data <= 95)]
print("Cleaned Data:",data)
```

##### Normalize

```python
normalized = (data - np.min(data)) / (np.max(data) - np.min(data))

print(normalized)
```

##### statistics

```python
print(np.mean(data))
print(np.median(data))
print(np.std(data))
print(np.min(data))
print(np.max(data))
```

##### sorting

```python
sort=np.sort(data)
print("Sorted data:", sort)
```

```python
def analyze_sensor(data):
    print("Mean:", np.mean(data))
    print("Max:", np.max(data))
    print("Min:", np.min(data))
    
    print("High values:", data[data > 70])
    
    print("Square roots:", np.sqrt(data))
    
    print("Top 5:", np.sort(data)[-5:])
analyze_sensor(data)
```
